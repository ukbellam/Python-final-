```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data from the CSV file
data = pd.read_csv('Input Data_Ass2_dataset.csv')
```

```python
#first 5 records
print("Displaying the first 5 records of the dataset:")
print(data.head())
```

```python
#the last 10 records
print("\nDisplaying the last 10 records of the dataset:")
print(data.tail(10))
```

```python
# Show the dimensions (rows and columns) of the data
print("\nThe dataset has the following dimensions (rows, columns):")
print(data.shape)
```

```python
# Get summary statistics for the 'age' column
print("\nSummary statistics for the 'age' column:")
print(data['age'].describe())
```

```python
# Selecting rows from index 150 to 249 for the training data
training_set = data.iloc[150:250]
# Display the selected rows
print("\nRows from index 150 to 249 stored in 'training_set':")
print(training_set)

```

```python
# Using the .loc method to select the 'bmi' and 'charges' columns
features = data.loc[:, ['bmi', 'charges']]

# Display the 'features' DataFrame
print("\n'features' DataFrame:")
print(features)
```

```python
# Selecting the last column using .iloc
labels = data.iloc[:, -1]

# Display the 'labels' DataFrame
print("\n'labels' DataFrame:")
print(labels)
```

```python
# Combining 'features' and 'labels' into one DataFrame using .join()
trainingData2F = features.join(labels)

# Display the concatenated 'trainingData2F'
print("\nConcatenated 'trainingData2F':")
print(trainingData2F)
```

```python
# Create a figure with custom size
plt.figure(figsize=(12, 8))

# Create the scatter plot with hue based on labels
sns.scatterplot(x=trainingData2F['bmi'], y=trainingData2F['charges'], hue=labels, palette='viridis')

# Add title and labels to the plot
plt.title('BMI vs Charges Scatter Plot')
plt.xlabel('BMI')
plt.ylabel('Charges')

# Display the legend with a custom title
plt.legend(title='Labels')

# Show the plot
plt.show()
```

```python
# Define the output filename
output_file = 'Ass2Output_BELLAMKONDA.csv'

# Save the DataFrame to a CSV file without the index
trainingData2F.to_csv(output_file, index=False)

# Print a message confirming the export
print(f"\nSuccessfully exported 'trainingData2F' to {output_file}")
```
